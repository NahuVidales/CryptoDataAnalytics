```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import data
```

```python
capitalizacion = pd.read_csv('data/capitalizaciones.csv')
capitalizacion.head()
```

```python
#En este caso solo nos interesa la columna de 'data.total_market_cap.usd' que nos devuelva un valor entero
capitalizacion['data.total_market_cap.usd'].astype(int)
```

```python
#Como se puede observar la capitalizacion del mercado crypto segun coingecko es de 1.22 billones de dolares lo que nos lleva a preguntarnos
#¿Que tan grande es el mercado crypto en comparacion con el mercado de acciones?
#Segun el sitio web de the balance el mercado de acciones tiene un valor de 95.71 billones de dolares
#Por lo que el mercado crypto es 78.5 veces mas pequeño que el mercado de acciones
#Fuente: https://www.indexmundi.com/facts/indicators/CM.MKT.LCAP.CD/rankings

#Lo veremos reglejado en un analisis de datos circular
# Datos
crypto_market_cap = 1.22  # En billones de dólares
stock_market_cap = 95.71  # En billones de dólares

# Etiquetas para el gráfico
labels = ['Crypto Market', 'Stock Market']

# Valores para el gráfico
sizes = [crypto_market_cap, stock_market_cap]

# Colores
colors = ['#66b3ff', '#99ff99']

# Crear el gráfico circular
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Título del gráfico
plt.title('Comparación entre Capitalización de Mercado Crypto y Mercado de Acciones')

# Mostrar el gráfico
plt.show()
```

```python
#Visualizando esto, podemos ver cuan superior es el volumen del mercado de bolsa sin embargo esto no quita que en algun momento una fraccion
#del mercado de bolsa pueda ser absorbido por el mercado de criptomonedas. ya que las operaciones de bolsa estan respaldadas por dolares americanos
#y como se ha estado viendo algunas naciones buscan sustituirlo por otras divisas, no estaria mal optar por una criptomoneda como el bitcoin
#para respaldar las operaciones de bolsa, ya que es una moneda que no esta respaldada por ningun gobierno y es descentralizada, por lo que no
#se puede manipular su valor, ademas de que es una moneda que no se puede falsificar, por lo que es una moneda muy segura para respaldar las operaciones
```

```python
#Primero veremos arkham donde en el eje y tenemos el precio y en el eje x tenemos el dia
arkham = pd.read_csv('data/prices_df_arkham.csv')
dias = arkham['dia']
precios = arkham['price']

# Crear el gráfico de líneas
plt.figure(figsize=(10, 6))
plt.plot(dias, precios, marker='o', linestyle='-', color='b', label='Precio de Arkham')

# Etiquetas y título
plt.xlabel('Día')
plt.ylabel('Precio')
plt.title('Precio de Arkham a lo largo del tiempo')
plt.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mayor claridad
plt.show()
```

```python
#Haremos lo mismo con ethereum y otras mas e intentaremos llegar a una conclusion
ethereum = pd.read_csv('data/prices_df_ethereum.csv')
plt.figure(figsize=(15, 5))
plt.plot(ethereum['dia'], ethereum['price'])
plt.title('Ethereum')
plt.xlabel('Dia')
plt.ylabel('Precio')
plt.show()
```

```python
#Ya se empieza a notar una diferencia entre los graficos de ambas criptomonedas
#Probaremos con el caso de Bitcoin
bitcoin = pd.read_csv('data/prices_df_bitcoin.csv')
plt.figure(figsize=(15, 5))
plt.plot(bitcoin['dia'], bitcoin['price'])
plt.title('Bitcoin precio por dia')
plt.xlabel('Dia')
plt.ylabel('Precio')
plt.show()
```

```python
#Como podemos observar bitcoin y ethereum tienen graficos similares, esto se debe a que el precio de ethereum depende del precio de bitcoin
#ya que la capitalizacion de mercado de cripto esta dada por un 50% de bitcoin ademas de ser los proyectos a los que mas recurren los inversores
#de largo plazo ya que a la hora de invertir estan entre esas 2 opciones
```
