### **Capitalizacion mercado cripto y bursatil**

Como se puede observar la capitalizacion del mercado crypto segun coingecko es de 1.22 billones de dolares lo que nos lleva a preguntarnos
¿Que tan grande es el mercado crypto en comparacion con el mercado de acciones?
Segun el sitio web de the balance el mercado de acciones tiene un valor de 95.71 billones de dolares
Por lo que el mercado crypto es 78.5 veces mas pequeño que el mercado de acciones
Fuente: https://www.indexmundi.com/facts/indicators/CM.MKT.LCAP.CD/rankings


```python

#Lo veremos reglejado en un analisis de datos circular
# Datos
crypto_market_cap = 1.22  # En billones de dólares
stock_market_cap = 95.71  # En billones de dólares

# Etiquetas para el gráfico
labels = ['Crypto Market', 'Stock Market']

# Valores para el gráfico
sizes = [crypto_market_cap, stock_market_cap]

# Colores
colors = ['#66b3ff', '#99ff99']

# Crear el gráfico circular
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Título del gráfico
plt.title('Comparación entre Capitalización de Mercado Crypto y Mercado de Acciones')

# Mostrar el gráfico
plt.show()
```

### **Conclusion analisis grafico**
Visualizando esto, podemos ver cuan superior es el volumen del mercado de bolsa sin embargo esto no quita que en algun momento una fraccion

### **Exchanges**
Los exchanges son el aparato financiero mediante el cual nosotros depositaremos nuestras divisas para que podamos cambiarlo por criptomonedas
Veremos tambien cuales son los exchanges mas importantes y cuales son los mas importantes en cuanto a volumen de transacciones ademas de cuales son los mas confiables ya que en ellos depositaremos nuestros activos y si depositamos nuestro dinero en uno y de no verificarlo antes podemos perder todo nuestro dinero

```python
exchanges = pd.read_csv('data/exchanges.csv')
exchanges.head(10)
```

Como podemos ver estan organizados en ranking de confianza y podemos observar como binance se posiciona en la parte superior de la lista
Por lo que sera nuestra mejor opcion para realizar la adquisicion de criptoactivos

- **Confianza del exchange:** 

Podemos ver tanto el volumen de transacciones en 24 horas como el trust score de cada exchange, por lo que podemos hacer una metrica entre ambos para ver que tan confiable es el exchange que estamos analizando y tomarlo como confiabilidad segun usuarios ya que el volumen en 24 horas que realizan los usuarios dentro de este es un buen indicador de que tan confiable es el exchange

```python
exchanges['confianza usuario'] = exchanges['trade_volume_24h_btc_normalized'] / 10000 * exchanges['trust_score']
exchanges.sort_values(by='confianza usuario', ascending=False).head(10)
```
